```python
import numpy as np
a = [[8,3,-3],[-2,-8,5], [3,5,10]]

diag = np.diag(np.abs(a))

off_diag = np.sum(np.abs(a), axis = 1)
if np.all(diag > off_diag):
  print('Matrix is diagonally dominant')
else:
  print('Matrix diagonally dominant')
```

```python
x1 = 0
x2 = 0
x3 = 0
epsilon = 0.01
converged = False

x_old = np.array([x1, x2, x3])

for k in range(1, 50):
  x1 = (14 - 3*x2 + 3*x3) / 8
  x2 = (5 + 2*x1 - 5*x3) / -8
  x3 = (-8 - 3*x1 - 5*x3) / 10
  x_new = np.array([x1, x2, x3])
  dx = np.sqrt(np.dot(x_new - x_old, x_new - x_old))

  print(f"{k}, {x1:.4f}, {x2:.4f}, {x3:.4f}")

  if dx < epsilon:
    converged = True
    print('Converged!')
    break
  x_old = x_new

if not converged:
  print('Not Converged!, increase iterations')
```

```python
import numpy as np

A = np.array([[4, 3, -5], [-2, -4, 5], [8,8,0]])
y = np.array([2, 5, -3])

x = np.linalg.solve(A,y)
print(x)
```

```python
import numpy as np

A = np.array([[4, 3, -5], [-2, -4, 5], [8,8,0]])
y = np.array([2, 5, -3])

A_inv = np.linalg.inv(A)
x = np.dot(A_inv, y)
print(x)
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0,1,101)
y = 1 + x + x * np.random.random(len(x))

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y,'b.', label='Data')
plt.plot(x, alpha[0]*x + alpha[1], 'r', label='Fit')
plt.xlabel('x')
plt.xlabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0,1,101)
y = 1 + x + x * np.random.random(len(x))

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

pinv = np.linalg.pinv(A)
alpha = pinv.dot(y)
print(alpha)
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0,1,101)
y = 1 + x + x * np.random.random(len(x))

A = np.vstack([x, np.ones(len(x))]).T

y = y[:, np.newaxis]

alpha = np.linalg.lstsq(A, y, rcond=None)[0]
print(alpha)
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x = np.linspace(0,1,101)
y = 1 + x + x * np.random.random(len(x))

def func(x, a, b):
  return a * x + b

alpha, _ = optimize.curve_fit(func, x, y)
print(alpha)
```
