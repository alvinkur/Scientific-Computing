```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [1,3,2]

f = interp1d(x,y)

y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1,5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from re import I
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x,y):
  n = len(y)
  coef = np.zeros([n,n]) #Membuat tabel kosong untuk divided differences
  coef[:, 0] = y #Kolom pertama bernilai nilai y

  for j in range(1, n): #Loop untuk mengisi kolom-kolom berikutnya
    for i in range(n - j):
      coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

  return coef

def newton_poly(coef, x_data, x):
  n = len(x_data) - 1
  p = coef[n]

  for k in range (1, n + 1):
    p = coef[n-k] + (x - x_data[n - k]) * p

  return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

#Menghitung koefisien divided difference
a_s = divided_diff(x,y)[0, :]

#Evaluasi interpolasi pada titik-titik baru
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

#Membuat plot
plt.figure(figsize=(12, 8), facecolor='#FAD0C4')
plt.plot(x, y, 'o', color='#0D4715') #Menampilkan titik data asli
plt.plot(x_new, y_new, color='#E9762B') #Menampilkan kurva interpolasi
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [1, 3, 2]

f = CubicSpline(x, y, bc_type='natural')

x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

plt.figure(figsize=(10,8), facecolor='#FAD0C4')
plt.plot(x_new, y_new, color='#A31D1D')
plt.plot(x, y, 'o', color='#003092')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
